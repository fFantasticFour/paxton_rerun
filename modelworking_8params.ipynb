Some function definitions:

* the model calls three different velocity dispersions, so it's cleaner to have a function for them
* the los distribution is a very cut-off gaussian, so renormalizing with erf is necessary
* the log likelihood, using flat priors whose limits are defined in a long 'if' statement.

```python
def calculate_vmax(richness):
    """Calculate the absolute velocity cutoff for interlopers in m/s.
    Follows equation (7) in redMaPPer IV.
    Parameters
    ----------
    richness : float
        Cluster richness, i.e. estimated number of galaxies belonging
        to the cluster.
    Returns
    -------
    vmax : float
        Maximum velocity value as set forth in equation (7) of redMaPPer IV.
        Units: 100 km/s
    """
    vmax = 3e6 * (richness / 20.0) ** 0.45 /100000 *10 # 100km/s
    return vmax  # m/s
```

```python
def background_gaussian(velocity, richness, total_sigma):
    """
    This function is used for evaluation of the line of sight
    gaussian. Sure, all Gaussians will have some fraction
    outside the vmax cut we made, and we'd have to normalize
    for that effect, but that amount is negligible for the
    orbiting and infalling Gaussians. But for the background
    Gaussian, a significant amount is cut off by the vmax cut,
    so I use an erf to scale and account for that."""
    vmax = 3e6 * (richness / 20.0) ** 0.45 / 100000 * 10
    outside_cut = np.where(velocity > vmax)[0]
    assert len(outside_cut) == 0, "UH OH: VELOCITY DATA FED INTO MCMC IS OUTSIDE CUT"

    z = vmax*(0.5**0.5)/total_sigma
    area_inside = erf(z)
    assert (area_inside != 0).all(), "Problem with vmax/sigma: " + str(vmax[np.where(z==0)]) + str(total_sigma[np.where(z==0)])
    normalization = 1/area_inside
    gaussian = np.exp(-0.5*((velocity/total_sigma)**2))/(np.sqrt(2*np.pi)*total_sigma)
    return_arr = normalization * gaussian

    if len(outside_cut) > 0:
        print("UH OH: VELOCITY DATA FED INTO MCMC IS OUTSIDE CUT")
        return_arr[outside_cut] = 0

    return return_arr
```

```python
def nlnlike_2pop(parameters):
    
    global v,r,rich,zcen,sigma_obs, min, max, nu
    
    a0,c0, sp_bg,alpha_inf, alpha_bg,sp_inf=parameters
    
    a1=0
    c1=-0.75
    
    z_p = np.median(zcen)
    lambda_p = np.median(rich)
    
    inf_sigmav_lam = sp_inf*(rich/lambda_p)**alpha_inf
    inf_sigmav = inf_sigmav_lam
    inf_sigmat = (inf_sigmav**2 + sigma_obs**2)**0.5
    #inf_dist = 1/np.pi/inf_sigmav/np.cosh(v/inf_sigmav)
    inf_dist = 1/2.0/inf_sigmav/np.cosh(np.pi/2.0*v/inf_sigmav)
    
    los_sigmav = sp_bg*(inf_sigmav_lam/sp_inf)**alpha_bg
    los_sigmat = (los_sigmav**2 + sigma_obs**2)**0.5
    los_dist = background_gaussian(v, rich, los_sigmav)

    r_lambda=(rich / 100.0) ** 0.2
    r_edge = 2.2*r_lambda
        
    A_inf = c0*(r/r_edge)**c1
    A_los = a0*(r/r_edge)**a1
    
    total = A_inf+A_los
    A_inf=A_inf/total
    A_los=A_los/total
    
    lp=0

    withinlimits=[]
    for i in range(len(parameters)):
        withinlimits=np.append( withinlimits, (min[i]<parameters[i]<max[i]))
    if np.all(withinlimits.astype(bool)):
        lp=0
    else:
        lp=-np.inf
        
    return -np.sum(np.log(A_inf*inf_dist+A_los*los_dist))-lp
```

```python
def lnlike_2pop(parameters):
    return -nlnlike_2pop(parameters)
```

```python
def plot_fit_2pop(df,parameters,minr,maxr,num_resample,fig,ax1):
    
    num_data_bin_edges = 50
    num_sim_bin_edges = 251
    min_x = -6
    max_x = 6

    sim_x_vals = np.linspace(min_x, max_x, num_sim_bin_edges)
    data_x_vals = np.linspace(min_x, max_x, num_data_bin_edges)
    data_bin_centers = (data_x_vals[:-1] + data_x_vals[1:])/2.
    data_bin_width = data_bin_centers[1] - data_bin_centers[0]
    sim_bin_centers = (sim_x_vals[:-1] + sim_x_vals[1:])/2.
    sim_bin_width = sim_bin_centers[1] - sim_bin_centers[0]
    
    z_p = np.median(df.zcen)
    lambda_p = np.median(df.rich)

    a0,c0, sp_bg,alpha_inf, alpha_bg,sp_inf=parameters
    
    a1=0
    c1=-0.75
    
    rich=df.rich
    zcen=df.zcen
    r=df.r
    v=df.v
    sigma_obs = df.sigma_obs
    r_lambda=(rich / 100.0) ** 0.2
    
    inf_sigmav_lam = sp_inf*(rich/lambda_p)**alpha_inf
    inf_sigmav = inf_sigmav_lam
    inf_sigmat = (inf_sigmav**2 + sigma_obs**2)**0.5
    r_edge = 2.2*r_lambda
    
    df2=df[ (df.r/r_edge>minr) & (df.r/r_edge <=maxr) ].copy()
    df2=df2.reset_index()

    rich=df2.rich
    zcen=df2.zcen
    r=df2.r
    v=df2.v
    sigma_obs = df2.sigma_obs
    r_lambda=(rich / 100.0) ** 0.2
    
    inf_sigmav_lam = sp_inf*(rich/lambda_p)**alpha_inf
    inf_sigmav = inf_sigmav_lam
    r_edge = 2.2*r_lambda
    
    #define scaling sigma_v
    sigmav_lam = sp_inf*(rich/lambda_p)**alpha_inf
    sigma_v = sigmav_lam
    
    A_inf = c0*(r/r_edge)**c1
    #assuming all are outside r_edge
    A_los = a0*(r/r_edge)**a1

    total = A_inf+A_los
    A_inf=A_inf/total
    A_los=A_los/total

    f_da=A_inf
    
    inf_v = []
    bg_v = []
    
    for i in tqdm(range(num_resample)):
        fraction = np.random.rand(len(v))
        inf = np.where(fraction < f_da)[0]
        los = np.where(fraction >= f_da)[0]

        inf_sigmav_lam = sp_inf*(rich[inf]/lambda_p)**alpha_inf
        inf_sigmav = inf_sigmav_lam
        inf_sigmat = (inf_sigmav**2 + sigma_obs[inf]**2)**0.5
        
        inf_v = np.append(inf_v,hypsecant.rvs(scale=inf_sigmav*2/np.pi)/sigma_v[inf])

        inf_sigmav_lam = sp_inf*(rich[los]/lambda_p)**alpha_inf
        inf_sigmav = inf_sigmav_lam
        inf_sigmat = (inf_sigmav**2 + sigma_obs[los]**2)**0.5
        bg_sigmav = sp_bg*(inf_sigmav_lam/sp_inf)**alpha_bg
        bg_sigmat = (bg_sigmav**2 + sigma_obs[los]**2)**0.5
        vmax = calculate_vmax(rich[los])
        bg_v=np.append(bg_v,truncnorm.rvs(-vmax/bg_sigmav,vmax/bg_sigmav, scale=bg_sigmav)/sigma_v[los])
   
    inf_count, bin_edges = np.histogram(inf_v, sim_x_vals)
    bg_count, bin_edges = np.histogram(bg_v, sim_x_vals)
    inf_normalized = inf_count / len(v) / sim_bin_width / num_resample
    bg_normalized = bg_count / len(v) / sim_bin_width / num_resample
    
    normalized_v=v/sigma_v
    
    counts, bin_edges = np.histogram(normalized_v, data_x_vals)
    normalized_counts = counts / len(v) / data_bin_width
    
    inf_lines = ax1.plot(sim_bin_centers, inf_normalized, linewidth=5, ls="dotted")
    los_lines = ax1.plot(sim_bin_centers, bg_normalized, linewidth=5, ls="dashed")
    sum_lines = ax1.plot(sim_bin_centers, inf_normalized + bg_normalized, linewidth=5)
    ax1.errorbar(data_bin_centers, normalized_counts, yerr=np.sqrt(counts)/data_bin_width/len(v), fmt='o',markersize=12,linewidth=5)
    ax1.set_xlim([-5, 5])
    ax1.set_ylim([-0.01, 0.6])
    ax1.set_xlabel('$v$ / $\sigma_{v,inf}$')
    ax1.set_ylabel('P($v$ / $\sigma_{v,inf})$')
    
    inset_ax = inset_axes(ax1,
                      height="30%", # set height
                      width="30%", # and width
                      loc='upper right')
    inset_ax.axvspan(minr, maxr, alpha=0.5, color='red')
    inset_ax.set_xlim(0,2.5)
    inset_ax.set_yticks([])
    inset_ax.set_xlabel('$R/R_{edge}$')  
```

```python
%%cython

from cpython.pycapsule cimport (PyCapsule_New,
                                PyCapsule_GetPointer)
from cpython.mem cimport PyMem_Malloc,  PyMem_Free
from libc.math cimport exp, sqrt
import scipy

cdef double c_integrand(double x, void* user_data):
    """The integrand, written in Cython"""
    # Extract a.
    # Cython uses array access syntax for pointer dereferencing!    
    cdef double R = (<double*>user_data)[0]
    cdef double alpha = (<double*>user_data)[1]
    cdef double r_s = (<double*>user_data)[2]
    cdef double r_t = (<double*>user_data)[2]
    
    r=x
    
    beta=4
    gamma=6
    arg=-2/alpha*((r/r_s)**alpha-1)
    coeff=(1+(r/r_t)**beta)**(-gamma/beta)
    rho_3d=exp(arg)*coeff
    
    return rho_3d*r/sqrt(r**2-R**2)

#
# Now comes some classic C-style housekeeping
#

cdef object pack_a(double R, double alpha, double r_s, double r_t):
    """Wrap 'a' in a PyCapsule for transport."""
    # Allocate memory where 'a' will be saved for the time being
    cdef double* a_ptr = <double*> PyMem_Malloc(4*sizeof(double))
    a_ptr[0] = R
    a_ptr[1] = alpha
    a_ptr[2] = r_s
    a_ptr[3] = r_t
    return PyCapsule_New(<void*>a_ptr, NULL, free_a)

cdef void free_a(capsule):
    """Free the memory our value is using up."""
    PyMem_Free(PyCapsule_GetPointer(capsule, NULL))

def get_low_level_callable(double R, double alpha, double r_s, double r_t):
    # scipy.LowLevelCallable expects the function signature to
    # appear as the "name" of the capsule
    func_capsule = PyCapsule_New(<void*>c_integrand,
                                 "double (double, void *)",
                                 NULL)
    data_capsule = pack_a(R,alpha,r_s,r_t)
    
    return scipy.LowLevelCallable(func_capsule, data_capsule)
```

```python
@np.vectorize
def proj2d_c(alpha,r_s,r_t,R,R_edge):
    c_integrand = get_low_level_callable(R,alpha,r_s,r_t)
    y, abserr = quad(c_integrand, R, R_edge)
    return 2*y
```

```python
def nlnlike6(parameters):
    
    global v,r,rich,zcen,sigma_obs, mymin, mymax, alpha, r_s,count, Surface
    
    #print(count)
    #count=count+1
    
    r_pedge,sp_inf,sp_orb,sp_bg,kt, r_pinf, r_porb, alpha_rinf=parameters
    
    a1=0 # from 2-pop fit outside r_edge
    c1=-.75 #From simulated fit
    a0=0.0160046096 # from 2-pop fit outside r_edge
    c0=0.00859426331 # from 2-pop fit outside r_edge
    b0=0.237 #From simulated fit
    b1=-6.11 #From simulated fit

    q_inf=1.637524 #From simulated fit
    alpha_inf=0.166803 #From simulated fit

    #sp_bg=20.008074 # from 2-pop fit outside r_edge
    alpha_bg=2.753028 # from 2-pop fit outside r_edge

    q_orb=2.8 #estimated from Han's figure
    alpha_orb=.5*7 #estimated from Han's figure

    alpha_redge=1.32#From Tomooka
    #alpha_rinf=3.308580 #From simulated fit
    alpha_rorb=alpha_rinf #A guess

    #kt=1.6
    
    z_p = np.median(zcen)
    lambda_p = np.median(rich)

    inf_sigmav_lam = sp_inf*(rich/lambda_p)**alpha_inf
        
    r_lambda=(rich / 100.0) ** 0.2
    #r_edge = 2.2*r_lambda
    r_edge = r_pedge*(inf_sigmav_lam/sp_inf)**alpha_redge

    r_inf = r_pinf*(inf_sigmav_lam/sp_inf)**alpha_rinf
    #Xinf=get_Xinf(rich, lambda_p, inf_sigmav_lam)
    #r_inf = r_edge * Xinf
    inf_sigmav = inf_sigmav_lam*(q_inf*np.exp(-r/r_inf)+1)
    inf_sigmat = (inf_sigmav**2 + sigma_obs**2)**0.5
    inf_dist = 1/2.0/inf_sigmav/np.cosh(np.pi/2.0*v/inf_sigmav)
    
    r_orb = r_porb*(inf_sigmav_lam/sp_inf)**alpha_rorb
#    Xorb=get_Xorb(rich, lambda_p, inf_sigmav_lam)
#    r_orb = r_edge * Xorb
    orb_sigmav_lam=sp_orb*(inf_sigmav_lam/sp_inf)**alpha_orb
    orb_sigmav = orb_sigmav_lam*(q_orb*np.exp(-r/r_orb)+1)
    orb_sigmat = (orb_sigmav**2 + sigma_obs**2)**0.5  
    #orb_dist = np.exp(-0.5*((v/orb_sigmav)**2))/(np.sqrt(2*np.pi)*orb_sigmav)
    orb_dist = 1/2.0/orb_sigmav/np.cosh(np.pi/2.0*v/orb_sigmav)
    
    los_sigmav = sp_bg*(inf_sigmav_lam/sp_inf)**alpha_bg
    #los_sigmav=sp_bg
    los_sigmat = (los_sigmav**2 + sigma_obs**2)**0.5
    los_dist = background_gaussian(v, rich, los_sigmav)
        
    A_inf = c0*(r/r_edge)**c1
    points=list(zip(alpha,r_s,r_edge/kt,r))
    A_orb=Surface.at( points )
    #A_orb = b0*np.exp(b1*r/r_edge)
    #A_orb=proj2d_c(alpha,r_s,r_edge/kt,r,r_edge)
    A_los = a0*(r/r_edge)**a1
    
    total = A_inf+A_los+A_orb
    A_inf=A_inf/total
    A_los=A_los/total
    A_orb=A_orb/total
    
    lp=0

    withinlimits=[]
    for i in range(len(parameters)):
        withinlimits=np.append( withinlimits, (mymin[i]<parameters[i]<mymax[i]))
    if np.all(withinlimits.astype(bool)):
        lp=0
    else:
        lp=-np.inf
        
    return -np.sum(np.log(A_inf*inf_dist+A_los*los_dist+A_orb*orb_dist))-lp
```

```python
def plot_fit(df,parameters,minr,maxr,num_resample,fig,ax1):
    
    num_data_bin_edges = 50
    num_sim_bin_edges = 251
    min_x = -6
    max_x = 6

    sim_x_vals = np.linspace(min_x, max_x, num_sim_bin_edges)
    data_x_vals = np.linspace(min_x, max_x, num_data_bin_edges)
    data_bin_centers = (data_x_vals[:-1] + data_x_vals[1:])/2.
    data_bin_width = data_bin_centers[1] - data_bin_centers[0]
    sim_bin_centers = (sim_x_vals[:-1] + sim_x_vals[1:])/2.
    sim_bin_width = sim_bin_centers[1] - sim_bin_centers[0]
    
    z_p = np.median(df.zcen)
    lambda_p = np.median(df.rich)

    r_pedge,sp_inf,sp_orb,sp_bg,kt, r_pinf, r_porb, alpha_rinf=parameters
    
    a1=0 # from 2-pop fit outside r_edge
    c1=-.75 #From simulated fit
    a0=0.0160046096 # from 2-pop fit outside r_edge
    c0=0.00859426331 # from 2-pop fit outside r_edge
    b0=0.237 #From simulated fit
    b1=-6.11 #From simulated fit

    q_inf=1.637524 #From simulated fit
    alpha_inf=0.166803 #From simulated fit

    #sp_bg=20.008074 # from 2-pop fit outside r_edge
    alpha_bg=2.753028 # from 2-pop fit outside r_edge

    q_orb=2.8 #estimated from Han's figure
    alpha_orb=.5*7 #estimated from Han's figure

    alpha_redge=1.32#From Tomooka
    #alpha_rinf=3.308580 #From simulated fit
    alpha_rorb=alpha_rinf #A guess
    
    #kt=1.6
    
    rich=df.rich
    zcen=df.zcen
    r=df.r
    v=df.v
    sigma_obs = df.sigma_obs
    r_lambda=(rich / 100.0) ** 0.2
    alpha=df.alpha
    r_s=df.r_s
    
    #r_edge = 2.2*r_lambda
    sigmav_lam = sp_inf*(rich/lambda_p)**alpha_inf
    sigma_v = sigmav_lam
    r_edge = r_pedge*(sigmav_lam/sp_inf)**alpha_redge
    
    df2=df[ (df.r/r_edge>minr) & (df.r/r_edge <=maxr) ].copy()
    df2=df2.reset_index()

    rich=df2.rich
    zcen=df2.zcen
    r=df2.r
    v=df2.v
    sigma_obs = df2.sigma_obs
    r_lambda=(rich / 100.0) ** 0.2
    alpha=df2.alpha
    r_s=df2.r_s
    
    #define scaling sigma_v
    sigmav_lam = sp_inf*(rich/lambda_p)**alpha_inf
    sigma_v = sigmav_lam
    
    #r_edge = 2.2*r_lambda
    #r_edge = r_pedge*(sigma_v/sp_inf)**alpha_redge
    r_edge = r_pedge*(sigmav_lam/sp_inf)**alpha_redge
    
    A_inf = c0*(r/r_edge)**c1
    A_los = a0*(r/r_edge)**a1
    points=list(zip(alpha,r_s,r_edge/kt,r))
    A_orb=Surface.at( points )
    #A_orb = b0*np.exp(b1*r/r_edge)
    #A_orb=proj2d_c(alpha,r_s,r_edge/kt,r,r_edge)
    
    total = A_inf+A_los+A_orb
    A_inf=A_inf/total
    A_los=A_los/total
    A_orb=A_orb/total

    f_da=A_inf+A_orb
    f_orb=A_orb/(A_inf+A_orb)

    orb_v = []
    inf_v = []
    bg_v = []
    
    for i in tqdm(range(num_resample)):
        fraction = np.random.rand(len(v))
        
        orb = np.where(fraction < f_da * f_orb)[0]
        inf = np.where((f_da * f_orb <= fraction) & (fraction < f_da))[0]
        los = np.where(fraction >= f_da)[0]
        
        inf_sigmav_lam = sp_inf*(rich[inf]/lambda_p)**alpha_inf

        r_inf = r_pinf*(inf_sigmav_lam/sp_inf)**alpha_rinf 
        #Xinf=get_Xinf(rich[inf], lambda_p, inf_sigmav_lam)
        #r_inf = r_edge[inf] * Xinf
        inf_sigmav = inf_sigmav_lam*(q_inf*np.exp(-r[inf]/r_inf)+1)
        inf_sigmat = (inf_sigmav**2 + sigma_obs[inf]**2)**0.5
        inf_v = np.append(inf_v,hypsecant.rvs(scale=inf_sigmav*2/np.pi)/sigma_v[inf])

        inf_sigmav_lam = sp_inf*(rich[orb]/lambda_p)**alpha_inf
        
        r_orb = r_porb*(inf_sigmav_lam/sp_inf)**alpha_rorb
        #Xorb=get_Xorb(rich[orb], lambda_p, inf_sigmav_lam)
        #r_orb = r_edge[orb] * Xorb
        orb_sigmav_lam=sp_orb*(inf_sigmav_lam/sp_inf)**alpha_orb
        orb_sigmav = orb_sigmav_lam*(q_orb*np.exp(-r[orb]/r_orb)+1)
        orb_sigmat = (orb_sigmav**2 + sigma_obs[orb]**2)**0.5
        #orb_v = np.append(orb_v,np.random.normal(0, orb_sigmav)/sigma_v[orb])
        orb_v = np.append(orb_v,hypsecant.rvs(scale=orb_sigmav*2/np.pi)/sigma_v[orb])
            
        inf_sigmav_lam = sp_inf*(rich[los]/lambda_p)**alpha_inf
        bg_sigmav = sp_bg*(inf_sigmav_lam/sp_inf)**alpha_bg
        bg_sigmat = (bg_sigmav**2 + sigma_obs[los]**2)**0.5
        vmax = calculate_vmax(rich[los])
        bg_v=np.append(bg_v,truncnorm.rvs(-vmax/bg_sigmav,vmax/bg_sigmav, scale=bg_sigmav)/sigma_v[los])
   
    orb_count, bin_edges = np.histogram(orb_v, sim_x_vals)
    inf_count, bin_edges = np.histogram(inf_v, sim_x_vals)
    bg_count, bin_edges = np.histogram(bg_v, sim_x_vals)
    orb_normalized = orb_count / len(v) / sim_bin_width / num_resample
    inf_normalized = inf_count / len(v) / sim_bin_width / num_resample
    bg_normalized = bg_count / len(v) / sim_bin_width / num_resample
    
    normalized_v=v/sigma_v
    
    counts, bin_edges = np.histogram(normalized_v, data_x_vals)
    normalized_counts = counts / len(v) / data_bin_width
    
    orb_lines = ax1.plot(sim_bin_centers, orb_normalized, linewidth=5, ls="dashdot")
    inf_lines = ax1.plot(sim_bin_centers, inf_normalized, linewidth=5, ls="dotted")
    los_lines = ax1.plot(sim_bin_centers, bg_normalized, linewidth=5, ls="dashed")
    sum_lines = ax1.plot(sim_bin_centers, inf_normalized + bg_normalized + orb_normalized, linewidth=5)
    ax1.errorbar(data_bin_centers, normalized_counts, yerr=np.sqrt(counts)/data_bin_width/len(v), fmt='o',markersize=12,linewidth=5)
    ax1.set_xlim([-8, 8])
    ax1.set_ylim([-0.01, 0.6])
    ax1.set_xlabel('$v$ / $\sigma_{v,inf}$')
    ax1.set_ylabel('P($v$ / $\sigma_{v,inf})$')
    
    inset_ax = inset_axes(ax1,
                      height="30%", # set height
                      width="30%", # and width
                      loc='upper right')
    inset_ax.axvspan(minr, maxr, alpha=0.5, color='red')
    inset_ax.set_xlim(0,2.5)
    inset_ax.set_yticks([])
    inset_ax.set_xlabel('$R/R_{edge}$')  
```
